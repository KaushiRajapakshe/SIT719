```python
import pandas as pd
import numpy as np
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
# Read the dataset in the iris.data file
iris = pd.read_csv(url)
# display
iris.head()
```

```python
# adding header 
header_list = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris.columns = header_list
iris.head()
```

```python
# Check for existing null values
total_null_values = sum(iris.isnull().sum())
print(total_null_values)

# Check for existing na values
total_na_values = sum(iris.isna().sum())
print(total_na_values)
```
